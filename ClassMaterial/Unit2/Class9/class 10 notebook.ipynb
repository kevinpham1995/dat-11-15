```python
from sklearn.pipeline import make_pipeline #allows you to chain encoding and modeling into single step
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree = DecisionTreeRegressor(max_depth = 5)
```

```python
pipe = make_pipeline(te, tree)
```

```python
pipe
```

```python
X = X.fillna(0)
```
