```python
# function to 
def create_word_matrix(sequence, embedding, embedding_dim=10):
    data_point = np.zeros((len(sequence), embedding_dim))
    for idx, val in enumerate(sequence):
        data_point[idx] = embedding[val]
        
    return data_point
```

```python
create_word_matrix(sequence1, embedding, 10)
```

```python
# wrap up our sequences in a list
sequences = [sequence1, sequence2]
```

```python
# final matrix to hold our embedding data
word_data = np.zeros((2, 8, 10))
```

```python
# fill in each row of the dataset with the encoded embedding data
for idx, row in enumerate(word_data):
    word_data[idx] = create_word_matrix(sequences[idx], embedding)
```

```python
# we'll now 'flatten' our dataset to get it ready for matrix multiplications
word_data = word_data.reshape(2, -1)

# and our new shape
word_data.shape
```

```python
# the dense layer for matrix multiplications
dense_layer = np.random.normal(0, .01, size=(80, 64))
```

```python
# a numpy array of weights
dense_layer
```

```python
# the relu activation to clip negative values to 0
def relu(X):
    return np.maximum(0, X)
```

```python
X_hidden1 = word_data @ dense_layer
# the value of X_hidden1
X_hidden1
```

```python
# pass it through ReLU to transform
X_hidden1 = relu(X_hidden1)
```

```python
# do it one more time for a second dense layer
dense_layer2 = np.random.normal(0, .01, size=(64, 64))
```

```python
# one more matrix multiplication for the second dense layer
X_hidden2 = X_hidden1 @ dense_layer2
# and pass it through relu activation
X_hidden2 = relu(X_hidden2)
```

```python
# we'll now create our sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# and the final layer
dense_layer3 = np.random.normal(0, .01, size = (64, 1))

# and our prediction
y_hat = X_hidden2 @ dense_layer3
# and apply the activation
y_hat = sigmoid(y_hat)
```

If we wanted to put the entire process into one code block, we could do it like the following:

```python
# this will create the forward pass to push our data through the neural network
y = np.array([[1, 0]])

# will do this for 5 epochs (training rounds)
for i in range(5):
    word_data = np.zeros((2, 8, 10))
    # load data into the word embedding
    for idx, row in enumerate(word_data):
        word_data[idx] = create_word_matrix(sequences[idx], embedding)
    # reshape it so it's two dimensional (the flatten layer)
    word_data = word_data.reshape(2, -1)
    # matrix multiply
    X_hidden1 = word_data @ dense_layer
    # relu activation
    X_hidden1 = relu(X_hidden1)
    # do it one more time
    X_hidden2 = X_hidden1 @ dense_layer2
    X_hidden2 = relu(X_hidden2)
    # and now create the output layer
    out = X_hidden2 @ dense_layer3
    # sigmoid to get the final probabilities
    y_hat = sigmoid(out)
    # you would calculate the final error at each round and use that to update the weights (the backward pass)
    gradient = y - y_hat
```

### Dealing With Sequential Dependency

So far, we've looked at two different layers:  the word embedding, which maps weights to individual words, and dense layers, which are just straight matrix multiplications.  

However, a unique aspect of neural networks is that you can use a variety of different neural network layers, and they allow you to connect one part of your data to the next in a different way, that can help you capture particular patterns in your data.  

Since we are dealing with natural language, an important level of detail we want to pay attention to is the **order in which words appear**.  Natural language is built to use words around it to convey meaning, but these details are not captured as well as they could be with our current setup.  

The next, more specialized layer that we're going to look at is something called a **recurrent neural network layer**.  It's frequently used for time series and natural language data because its calculation technique naturally captures sequential detail.  

Here's how it works:

```python
# the forward pass for a recurrent neural network
timesteps = 8 # number of samples in our sequence
input_features = 10 # number of columns of weights for each sample
output_features = 32 # number of columns each sample will have at the end

# create a bunch of weights for multiplications -- because that's what Neural Networks do
W = np.random.random((output_features, input_features))
U = np.random.random((output_features, output_features))
b = np.random.random((output_features, ))

# material we are going to use to store data after computing the 
# entire layer for every sample
layer_outputs = []

# recreate our word embedding
word_data = np.random.random((2, 8, 10))

for sample in word_data:
    state_t = np.zeros((output_features, ))
    # will contain sequential data after each step / sample
    sample_outputs = []
    for word in sample:
        # for each word, do these matrix operations
        output_t = np.tanh(W @ word) + U @ state_t + b
        # append the end
        sample_outputs.append(output_t)
        # set the new state previous to the last state
        state_t = output_t
    layer_outputs.append(sample_outputs)
```
